```python
############# WRITE THE CODE IN THIS CELL ####################
# Removing punctuation method 1
def remove_punctuation(text):
    final = "".join(u for u in text if u not in ("?", ".", ";", ":",  "!",'"'))
    return final

#Applying the remove_punctuation function to the tweet_lowercase column of the DataFrame df
df['tweet_lowercase'] = df['tweet_lowercase'].apply(remove_punctuation)
```

```python
############# WRITE THE CODE IN THIS CELL ####################
#Removing punctuation method 2
import string

#Accessing the predefined set of punctuation characters 
string.punctuation

#Applying lambda function to remove punctuation characters from 'tweet_lowercase' column
df['tweet_lowercase']=df['tweet_lowercase'].apply(lambda x:''.join(i for i in x if i not in string.punctuation))
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# Stopwords
#Importing the nltk library
import nltk

#Importing the stopwords module from the nltk.corpus package
from nltk.corpus import stopwords

#Retrieving all English stopwords from the nltk.corpus package
allstopwords = stopwords.words('english')

#Applying a lambda function to each element in the 'tweet_lowercase' column 
df['tweet_lowercase']=df.tweet_lowercase.apply(lambda x: " ".join(i for i in x.split() if i not in allstopwords))
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# Extracting input and output
X=df['tweet_lowercase']
y=df['sentiment']

df
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# count vectorizer:
#Importing count vectorizer class from the sklearn.feature_extraction.text module
from sklearn.feature_extraction.text import CountVectorizer

#Initializing the count vectorizer with a custom token pattern that matches word boundaries
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')

#Transforming the text data X into a document-term matrix
X = vectorizer.fit_transform(X)
```

```python
############# WRITE THE CODE IN THIS CELL ####################
#Importing the train_test_split function from the sklearn.model_selection module
from sklearn.model_selection import train_test_split

#Splitting the dataset into training and testing sets
# X: Features (independent variables)
# y: Target (dependent variable)
# test_size: specifies the proportion of the dataset to include in the testing set
# random_state: controls the randomness of the split to ensure reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# Step 1: defining the classification models
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

# Initializing classifiers
SVM = svm.SVC()
RF = RandomForestClassifier()
KNN = KNeighborsClassifier()
DT=DecisionTreeClassifier()
NB = GaussianNB()
LR = LogisticRegression()
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# Step 2: training the models
# Fitting the classifiers to the training data
SVM.fit(X_train, y_train)
RF.fit(X_train, y_train)
KNN.fit(X_train, y_train)
DT.fit(X_train, y_train)
LR.fit(X_train,y_train)
NB.fit(X_train.toarray(),y_train)
```

```python
############# WRITE THE CODE IN THIS CELL ####################
#Step 3: prediction
y_pred1=SVM.predict(X_test)
y_pred2=RF.predict(X_test)
y_pred3=KNN.predict(X_test)
y_pred4=DT.predict(X_test)
y_pred5=LR.predict(X_test)
y_pred6=NB.predict(X_test.toarray())
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# This function takes the confusion matrix (cm) from the cell above and produces all evaluation matrix
def confusion_metrics (conf_matrix):

    TP = conf_matrix[1][1]
    TN = conf_matrix[0][0]
    FP = conf_matrix[0][1]
    FN = conf_matrix[1][0]
    print('True Positives:', TP)
    print('True Negatives:', TN)
    print('False Positives:', FP)
    print('False Negatives:', FN)

    # calculate accuracy
    conf_accuracy = (float (TP+TN) / float(TP + TN + FP + FN))

    # calculate mis-classification
    conf_misclassification = 1- conf_accuracy

    # calculate the sensitivity
    conf_sensitivity = (TP / float(TP + FN))
    # calculate the specificity
    conf_specificity = (TN / float(TN + FP))

    # calculate precision
    conf_precision = (TN / float(TN + FP))
    # calculate f_1 score
    conf_f1 = 2 * ((conf_precision * conf_sensitivity) / (conf_precision + conf_sensitivity))
    print('-'*50)
    print(f'Accuracy: {round(conf_accuracy,2)}')
    print(f'Mis-Classification: {round(conf_misclassification,2)}')
    print(f'Sensitivity: {round(conf_sensitivity,2)}')
    print(f'Specificity: {round(conf_specificity,2)}')
    print(f'Precision: {round(conf_precision,2)}')
    print(f'f_1 Score: {round(conf_f1,2)}')
```

```python
############# WRITE THE CODE IN THIS CELL ####################
# Creating the confusion matrics for all classifiers' predictions
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Calculating confusion matrix and creating ConfusionMatrixDisplay for SVM classifier
cm1 = confusion_matrix(y_test, y_pred1, labels=SVM.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm1,display_labels=SVM.classes_)
disp.plot()
plt.title("SVM")

# Calculating confusion matrix and creating ConfusionMatrixDisplay for RF classifier
cm2 = confusion_matrix(y_test, y_pred2, labels=RF.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm2,display_labels=RF.classes_)
disp.plot()
plt.title("RF")

# Calculating confusion matrix and creating ConfusionMatrixDisplay for KNN classifier
cm3 = confusion_matrix(y_test, y_pred3, labels=KNN.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm3,display_labels=KNN.classes_)
disp.plot()
plt.title("KNN")

# Calculating confusion matrix and creating ConfusionMatrixDisplay for DT classifier
cm4 = confusion_matrix(y_test, y_pred4, labels=DT.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm4,display_labels=DT.classes_)
disp.plot()
plt.title("DT")

# Calculating confusion matrix and creating ConfusionMatrixDisplay for LR classifier
cm5 = confusion_matrix(y_test, y_pred5, labels=DT.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm5,display_labels=DT.classes_)
disp.plot()
plt.title("LR")

# Calculating confusion matrix and creating ConfusionMatrixDisplay for NB classifier
cm6 = confusion_matrix(y_test, y_pred6, labels=NB.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm6,display_labels=NB.classes_)
disp.plot()
plt.title("NB")

#Displaying the confusion matrices for all classifiers
plt.show()
```

```python
############# WRITE THE CODE IN THIS CELL ####################
#printing the evaluation metrics for all classifiers
print('SVM metrics\n')
confusion_metrics(cm1)
print('\n\n')
print('RF metrics\n')
confusion_metrics(cm2)
print('\n\n')
print('KNN metrics\n')
confusion_metrics(cm3)
print('\n\n')
print('DT metrics\n')
confusion_metrics(cm4)
print('\n\n')
print('LR metrics\n')
confusion_metrics(cm5)
print('\n\n')
print('NB metrics\n')
confusion_metrics(cm6)
print('\n\n')
```
