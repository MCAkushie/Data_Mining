```python
############### Write your code in this cell (If applicable) ##################
# Importing the relevant libraries
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Reading csv file into a pandas DataFrame
basket_Portugal = pd.read_csv('C:/Users/HP/Downloads/Portugal_online_retail.csv')

#printing data
basket_Portugal
```

```python
#dropping the Invoice No Column
basket_Portugal=basket_Portugal.drop('InvoiceNo',axis=1)

#printing data to check that invoice no is dropped
basket_Portugal
```

```python
# Finding the frequent itemsets
frq_items = apriori(basket_Portugal, min_support = 0.15, use_colnames = True)

#Sorting and printing the values in descending order
frq_items.sort_values(by=['support'], ascending = False)
```

```python
#printing shape 
frq_items.shape
```

```python
# Collecting the inferred rules and defining a confidence level of 60%
rules = association_rules(frq_items, metric = 'confidence', min_threshold = 0.6)

#sorting and printing values in descending order of confidence level
rules.sort_values(by=['confidence'], ascending = False)
```

```python
# Collecting the inferred rules and defining a confidence level of 70%
rules = association_rules(frq_items, metric = 'confidence', min_threshold = 0.7)

#Sorting and printing rules by confidence in descending order 
rules.sort_values(by=['confidence'], ascending = False)
```

```python
# Collecting the inferred rules and defining a confidence level of 80%
rules = association_rules(frq_items, metric = 'confidence', min_threshold = 0.8)

#Sorting and printing rules by confidence in descending order 
rules.sort_values(by=['confidence'], ascending = False)
```
